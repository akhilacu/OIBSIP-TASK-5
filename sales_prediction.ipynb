```python
# Model Preparation

X = df.drop('Sales', axis=1)
y = df[["Sales"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=46)
```

```python
lin_model = sm.ols(formula="Sales ~ TV + Radio + Newspaper", data=df).fit()
```

```python
print(lin_model.params, "\n")
```

```python
print(lin_model.summary())
```

```python
# Define a list of models to evaluate
models = [('LinearRegression', LinearRegression())]
```

```python
for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f"{name}: RMSE = {rmse:.2f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df'
# Replace 'df' with your actual DataFrame

# Calculate the correlation matrix
corrmat = df.corr()

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 9))

# Iterate through each pair of variables and create box plots
for i in range(len(corrmat.columns)):
    for j in range(i + 1, len(corrmat.columns)):
        x_var = corrmat.columns[i]
        y_var = corrmat.columns[j]

        # Create a subplot for each pair
        plt.subplot(len(corrmat.columns), len(corrmat.columns), i * len(corrmat.columns) + j + 1)

        # Create box plots for each pair
        sns.boxplot(x=df[x_var], y=df[y_var])

        # Set axis labels
        plt.xlabel(x_var)
        plt.ylabel(y_var)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
